EX-1: Preparing the dataset.

Extract entrees corresponding to the Internet sales channel (excluding Mobile ones) used for a Round Trip (excluding Circle and One-Way Trips).

Load csv into dataframe and keep record of 
    1. size - number of passengers
    2. lead_time - purchase lead time
    3. booking - flag indicating if the booking was completed

```python
initial_df = pd.read_csv('customer_booking.csv', encoding='ISO-8859-1')
initial_df
```

```python
filtered_df = initial_df[(initial_df['sales_channel'] == 'Internet') & (initial_df['trip_type'] == 'RoundTrip')]
filtered_df
```

```python
filtered_df = filtered_df[['num_passengers', 'purchase_lead', 'booking_complete']]
filtered_df
```

```python
numpy_df = filtered_df.to_numpy()
numpy_df
```
